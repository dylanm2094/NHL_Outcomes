```python
import pandas as pd
```

```python
stats = 'https://www.hockey-reference.com/leagues/NHL_2020.html#all_stats'
```

```python
tables = pd.read_html(stats)
stats_df_1 = tables[0]
stats_df_1 = stats_df_1.drop([0, 9])
stats_df_1
```

```python
stats_df_2 = tables[1]
stats_df_2 = stats_df_2.drop([0, 8])
stats_df_2
```

```python
frames = [stats_df_1, stats_df_2]
df = pd.concat(frames)
df
```

```python
df = df.rename(columns={"Unnamed: 0": "Team","W": "season_wins"})
cols = [1,3,4,5,6,9,10,11,12,13,14]
df.drop(df.columns[cols],axis=1,inplace=True)
df
```

```python
net_g = []
for i in df.iterrows():
    #net_goals calculation
    net_g.append(int(i[1]['GF']) - int(i[1]['GA']))

df['net_goals'] = net_g
df
```
